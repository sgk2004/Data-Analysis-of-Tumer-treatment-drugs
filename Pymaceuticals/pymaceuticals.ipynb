```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the csv file - clinicaltrial_data.csv
file_path1= os.path.join('data','clinicaltrial_data.csv')
clinical_trial_df = pd.read_csv(file_path1)
```

```python
#Read the csv file - mouse_drug_data.csv
file_path2= os.path.join('data','mouse_drug_data.csv')
mouse_drug_df = pd.read_csv(file_path2)
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(clinical_trial_df, mouse_drug_df, on=('Mouse ID'))
#merged_df.info()
# Display the data table for preview
#merged_df.head()
```

## Tumor Response to Treatment

Find the Mean  of Tumor Volume(mm3) Data, Grouped by Drug and Timepoint 

```python
# Mean of Tumor Volume(mm3) Data, Grouped by Drug and Timepoint
mean_tv_df = pd.DataFrame(merged_df.groupby(['Drug', 'Timepoint']) ["Tumor Volume (mm3)"].mean())
#Reindexing DF
mean_tv_df=mean_tv_df.reset_index()
#preview
#mean_tv_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

mean_df_pivot=mean_tv_df.pivot(index='Timepoint',columns='Drug', values='Tumor Volume (mm3)')

#preview
mean_df_pivot
```

Analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare.

Mean of Tumor Volume 

```python
#Capomulin:Round the mean to two decimal point and make a list of mean of Tumor Volume 
cap_tv_list=mean_df_pivot['Capomulin'].to_list()
cap_tv_list=[round(item, 2) for item in cap_tv_list]
```

```python
#Ketapril:Round the mean to two decimal point and make a list of mean of Tumor Volume 
ket_tv_list=mean_df_pivot['Ketapril'].to_list()
ket_tv_list=[round(item, 2) for item in ket_tv_list]
```

```python
# Infubinol :Round the mean to two decimal point and make a list of mean of Tumor Volume 
inf_tv_list=mean_df_pivot['Infubinol'].to_list()
inf_tv_list=[round(item, 2) for item in inf_tv_list]
```

```python
#Placebo:Round the mean to two decimal point and make a list of mean of Tumor Volume 
pla_tv_list=mean_df_pivot['Placebo'].to_list()
pla_tv_list=[round(item, 2) for item in pla_tv_list]
```

Find  Standard Error of Mean (sem) of Tumor Volume(mm3) Data, Grouped by Drug and Timepoint 

```python
# Standard Error of Tumor Volume(mm3) Data, Grouped by Drug and Timepoint
sem_tv_df = pd.DataFrame(merged_df.groupby(['Drug', 'Timepoint']) ["Tumor Volume (mm3)"].sem())
#Reindexing DF
sem_tv_df=sem_tv_df.reset_index()
#preview
#sem_tv_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

sem_df_pivot=sem_tv_df.pivot(index='Timepoint',columns='Drug', values='Tumor Volume (mm3)')
sem_df_pivot=sem_tv_df.pivot(index='Timepoint',columns='Drug', values='Tumor Volume (mm3)')
#preview
sem_df_pivot
```

Standard Error of Mean(sem) of Tumor Volume of four treatments (Capomulin, Infubinol, Ketapril, and Placebo).

```python
#Capomulin :Round the Standard Error to two decimal point and make a list of SEM of Tumor Volume  
cap_sem_list=sem_df_pivot['Capomulin'].to_list()
#round the floating number to two decimal value
cap_sem_list=[round(item, 2) for item in cap_sem_list]
#cap_sem_list
```

```python
#Ketapril :Round the Standard Error to two decimal point and make a list of SEM of Tumor Volume 
ket_sem_list=sem_df_pivot['Ketapril'].to_list()
#round the floating number to two decimal value
ket_sem_list=[round(item, 2) for item in ket_sem_list]
#ket_sem_list
```

```python
#Infubinol:Round the Standard Error to two decimal point and make a list of SEM of Tumor Volume 
inf_sem_list=sem_df_pivot['Infubinol'].to_list()
#round the floating number to two decimal value
inf_sem_list=[round(item, 2) for item in inf_sem_list]
#inf_sem_list
```

```python
#Ketapril :Round the Standard Error to two decimal point and make a list of SEM of Tumor Volume 
pla_sem_list=sem_df_pivot['Placebo'].to_list()
#round the floating number to two decimal value
pla_sem_list=[round(item, 2) for item in pla_sem_list]
#pla_sem_list
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

days= np.arange(0, 50, 5)

#horzondal black line at starting tumer size
plt.hlines(45,0, max(days)+3, color='black')

ax.errorbar(days,cap_tv_list,yerr=cap_sem_list, fmt="red", label="Capomulin", marker='*')

ax.errorbar(days,ket_tv_list,yerr=ket_sem_list, fmt="black", label="Ketapril", marker= '^')

ax.errorbar(days,inf_tv_list,yerr=inf_sem_list, fmt="blue",  label="Infubinol", marker ='o')

ax.errorbar(days,pla_tv_list,yerr=pla_sem_list, fmt="green",  label="Placebo", marker ='s')

ax.legend(loc='upper center')

ax.set_xlabel('Time-Days')
ax.set_ylabel('Mean of Tumor Volume(mm3) with SEM')
ax.set_title('Tumor Response to Treatment')

tick_locations = [value for value in days]
ax.set_xticks(tick_locations, minor=False)
yvalue =np.arange(35,71,5)
yticks_location= [y1 for y1 in yvalue]
ax.set_yticks(yticks_location, minor=False)

plt.xlim(0, max(days)+3)
plt.grid()
#plt.figure(figsize= (100,1))

plt.tight_layout()

# Save the Figure
plt.savefig(os.path.join("Images","Tumor_response1"))
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint 
# Convert to DataFrame
mouse_count_df = pd.DataFrame(merged_df.groupby(['Drug', 'Timepoint']) ['Timepoint'].count())

#Rename column
mouse_count_df=mouse_count_df.rename(columns={'Timepoint':'Mouse Count'})

#Reindexing DF
mouse_count_df=mouse_count_df.reset_index()

#Display dataframe
mouse_count_df
```

SURVIVED MOUSE COUNT

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot=mouse_count_df.pivot(index='Timepoint',columns='Drug', values='Mouse Count')

# Preview the Data Frame
mouse_count_pivot
```

```python
#Make a list of number of mouse survived from the above DataFrame
#For drugs Capomulin, Infubinol, Ketapril, and Placebo

#Capomulin 
cap_mouse_list= mouse_count_pivot ['Capomulin'].to_list()
#mouse survival rate with drug Capomulin
cap_survrate = [(count/cap_mouse_list[0])*100 for count in cap_mouse_list]

#Ketapril 
ket_mouse_list=mouse_count_pivot['Ketapril'].to_list()
#mouse survival rate with drug Ketapril
ket_survrate = [(count/ket_mouse_list[0])*100 for count in ket_mouse_list]

# Infubinol 
inf_mouse_list=mouse_count_pivot['Infubinol'].to_list()
#mouse survival rate with drug Infubinol
inf_survrate = [(count/inf_mouse_list[0])*100 for count in inf_mouse_list]

#Placebo 
pla_mouse_list=mouse_count_pivot['Placebo'].to_list()
#mouse survival rate with drug Placebo
pla_survrate = [(count/pla_mouse_list[0])*100 for count in pla_mouse_list]

#Display the lists
# print(cap_mouse_list)
# print(cap_survrate)
# print(ket_mouse_list)
# print(ket_survrate)
# print(inf_mouse_list)
# print(inf_survrate)
# print(pla_mouse_list)
# print(pla_survrate)
```

```python
fig, ax = plt.subplots()

#Days on x axis
x_axis = np.arange(0, 50, 5)

# Generate the Plot (Accounting for percentages)
ax.plot(x_axis, cap_survrate,marker='*',color='red',label="Capomulin")
ax.plot(x_axis, ket_survrate,marker='^',color='black',label="Ketapril")
ax.plot(x_axis, inf_survrate,marker='o',color='blue',label="Infubinol")
ax.plot(x_axis, pla_survrate,marker='s',color='green',label="Placebo")

#Set labels
ax.set_xlabel('Time-Days')
ax.set_ylabel('Mouse survival rate (%)')
ax.set_title('Survival during treatment')

#display legend without blocking the data
ax.legend(loc='bottom left')

#Add grid or better  analysis
ax.grid(True)

fig.tight_layout()

#Display the graph
plt.show()

# Save the Figure
plt.savefig(os.path.join("Images","Survival_rate"))

```

```python
# Store all Relevant drugs (Capomulin, Infubinol, Ketapril, and Placebo) Percent Changes into a Tuple
y_axis= (cap_tv_change,inf_tv_change,ket_tv_change,pla_tv_change)

x_axis = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

# #Display results for checking
# print(type(y_axis))
#print((y_axis))
# print(x_axis)
```

```python
fig, ax = plt.subplots()

# Splice the data between passing and failing drugs
colors = []
for item in y_axis:
    if item < 0:
        colors.append('green')
    else:
        colors.append('red')

# Orient widths. Add labels, tick marks, etc. 
# Set the limit of the x and y axes. lim(start, end)
plt.xlim(-1, len(x_axis))
plt.ylim(min(y_axis)-5, max(y_axis)+10)

tumor_growth_bar = ax.bar(x_axis, y_axis, color=colors, alpha=1.0, edgecolor='black', linewidth=0.7)

ax.set_title("Tumor Change Over 45 Day Treatment")
ax.set_ylabel("% Tumor Volume Change")
ax.set_xlabel("Drugs")

xtick_locations = [value for value in x_axis]
ax.set_xticks(xtick_locations, minor=False )

yvalue =np.arange(-20,70,10)
yticks_location= [y2 for y2 in yvalue]
ax.set_yticks(yticks_location, minor=False)


# Plot horizontal line at y=0
plt.hlines(0, -1, len(x_axis), alpha=1.0, linewidth=0.5)

# Add gridlines
plt.grid('on', which='major', axis='y', linestyle='dotted', linewidth=0.5)

# Use functions to label the percentages of changes
def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

        # To fit the label above the column
        # otherwise, put it inside the column.
        if p_height > 0.95: # arbitrary; 95% looked good to me.
            label_position = height - (y_height * 0.05)
        else:
            label_position = height + (y_height * 0.01)

        ax.text(rect.get_x() + rect.get_width()/2., label_position,
                '{}%'.format('%d' % int(height)),
                ha='center', va='bottom')

# Call functions to implement the function calls
autolabel(tumor_growth_bar,ax)

# Save the Figure
plt.savefig(os.path.join("Images","Tumor_volume_change"))

fig.tight_layout()

# Show the Figure
plt.show()
```
